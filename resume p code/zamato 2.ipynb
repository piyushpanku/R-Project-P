```python
# Encoding categorical columns using OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder
OE = OrdinalEncoder()
for i in merged_data.columns:
    if merged_data[i].dtypes=='object':
        merged_data[i] = OE.fit_transform(merged_data[i].values.reshape(-1,1))
```

```python
x=merged_data.iloc[:,:-1]
y=merged_data.iloc[:,-1]
```

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score 
```

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y, random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print("At random state",i,"The model performs very well")
        print("At random_state:",i)
        print("Testing accuracy is:",accuracy_score(y_train,pred_train)*100)
        print("Testing accuracy is:",accuracy_score(y_test,pred_test)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
RFC = RandomForestClassifier()
RFC.fit(x_train,y_train)
predRFC = RFC.predict(x_test)
print(accuracy_score(y_test,predRFC))
print(confusion_matrix(y_test,predRFC))
print(classification_report(y_test,predRFC))
```

### save the pickle file

```python
import pickle
filename = 'zomato2.pkl'
pickle.dump(predRFC, open(filename, 'wb'))
```
